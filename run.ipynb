# Chat with Meta's Llama 3 8B

```python
import client

def chatbot(prompt, context=None):
    print(f"\n[User] {prompt}\n")
    print("[Agent]", end="")
    return client.chat("llama3", prompt, context=context)
```

```python
context = chatbot("Hey, please code me a softmax from scratch in Python")
```

```python
context = chatbot("Can you code it as a Triton kernel now?", context=context)
```

```python
chatbot("Thank you!", context=context)
```
